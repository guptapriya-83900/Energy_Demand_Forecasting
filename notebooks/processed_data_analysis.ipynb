```python
import pandas as pd
```

```python
# Path to the cleaned data file
cleaned_data_path = "../data/processed/cleaned_energy_data.csv"

# Load the cleaned data
df_cleaned = pd.read_csv(cleaned_data_path)
```

```python
# Display basic info about the cleaned data
print("Cleaned Data Overview:")
print(df_cleaned.info())
```

```python
# Preview the first few rows
print("\nFirst 5 Rows of Cleaned Data:")
print(df_cleaned.head())
```

```python

# Check for missing values
print("\nMissing Values in Cleaned Data:")
print(df_cleaned.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(df_cleaned.describe())
```

```python
print("Negative values in 'value':", (df_cleaned['value'] < 0).sum())
```

```python
print("Outliers in 'value':", (df_cleaned['value'] > 120000).sum())
```

```python
print("\nUnique values in 'hour':", df_cleaned['hour'].unique())
print("Unique values in 'weekday':", df_cleaned['weekday'].unique())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['value'], bins=50, color='blue', edgecolor='black')
plt.title("Distribution of Energy Demand (Cleaned Data)")
plt.xlabel("Energy Demand (Megawatthours)")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_cleaned['hour'], df_cleaned['value'], alpha=0.6, label="Hourly Demand")
plt.title("Hourly Energy Demand (Cleaned Data)")
plt.xlabel("Hour")
plt.ylabel("Energy Demand (Megawatthours)")
plt.grid()
plt.legend()
plt.show()
```
